```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate first datasets

N = 1000
X1 = np.random.normal(6, 1, N)

# Create an array of indices for the y-coordinate
indices = np.arange(len(X1))

# compare pdf and cdf of pressure
# outcomes
plt.scatter(indices, X1, marker='o', s=20, color='white', edgecolor='black')
plt.plot(np.arange(1, N + 1), np.linspace(6, 6, N), linewidth=1, color='black')
plt.xlabel(r'Outcome $\omega$', fontsize=14)
plt.ylabel(r'$x_\omega$ [bar]', fontsize=14)
plt.grid(True)
plt.axis([1, N, 0, 15])

plt.show()
```

```python
# Can also automatically visualize empirical pdf using histogram

# Create the histogram
plt.hist(X1, bins=np.arange(0, 15.1, 0.1), density=True, edgecolor='black', facecolor='none', linewidth=1)

# Customize the plot
plt.xlabel('$x$ [bar]', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.axis([0, 15, 0, 0.5])
plt.grid(False)
plt.show()
```

```python
# Compute an plot empirical CDF
dx = 0.1
t = np.arange(0,15+dx,dx)
N = len(t)

# Initialize the cumulative distribution function array
F = np.zeros(len(t))

# Compute the empirical CDF
for i in range(len(t)):
    for j in range(N):
        if X1[j] <= t[i]:
            F[i] += 1
# Normalize F
F = F/N
```

```python
# Plot the empirical CDF
plt.step(t, F, color='black', where='mid')
plt.xlabel('$x$ [bar]')
plt.ylabel('Frequency')
plt.grid(True)
plt.axis([0, 15, 0, 1.05])
plt.show()
```
