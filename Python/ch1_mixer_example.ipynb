```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate data for three flows
np.random.seed(0) # to ensure reproducibility
N = 100
time = np.arange(0, N, 1)
F1 = np.random.normal(100, 100*0.01/2.5, N)
F2 = np.random.normal(50, 50*0.01/2.5, N)
F3 = np.random.normal(150, 150*0.01/2.5, N)

# Compute conservation error
err = F3 - F2 - F1
```

```python
# Visualize flows
plt.scatter(time, F1, marker='o', s=20, color='white', edgecolor='black')
plt.plot(time, 100 * 1.01 * np.ones(N), 'k--')
plt.plot(time, 100 * 0.99 * np.ones(N), 'k--')
plt.xlabel('Observation $\omega$')
plt.ylabel('$f_1(\omega)$')
plt.grid(True)
plt.axis([0, N, 98, 102])
plt.show()
```

```python
# Visualize flows
plt.scatter(time, F2, marker='o', s=20, color='white', edgecolor='black')
plt.plot(time, 50 * 1.01 * np.ones(N), 'k--')
plt.plot(time, 50 * 0.99 * np.ones(N), 'k--')
plt.xlabel('Observation $\omega$')
plt.ylabel('$f_2(\omega)$')
plt.grid(True)
plt.axis([0, N, 48, 52])
plt.show()
```

```python
# Visualize flows
plt.scatter(time, F3, marker='o', s=20, color='white', edgecolor='black')
plt.plot(time, 150 * 1.01 * np.ones(N), 'k--')
plt.plot(time, 150 * 0.99 * np.ones(N), 'k--')
plt.xlabel('Observation $\omega$')
plt.ylabel('$f_3(\omega)$')
plt.grid(True)
plt.axis([0, N, 148, 152])
plt.show()
```

```python
# Visualize conservation error
plt.scatter(time, err, marker='o', s=20, color='white', edgecolor='black')
plt.plot(time, 0 * np.ones(N), 'k--')
plt.xlabel('Observation $\omega$')
plt.ylabel('$\epsilon$[$\omega$]')
plt.grid(True)
plt.show()
```
