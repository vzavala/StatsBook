```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
np.random.seed(1) # to ensure reproducibility
mu = 0
sigma = 1
x = np.linspace(-5, 5, 1000)

# Plot PDF
plt.plot(x, norm.pdf(x, mu, sigma), 'black', linewidth=1.5)

plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.grid(True)
plt.axis([-5, 5, 0, 0.42])
plt.show()
```

```python
# Plot CDF
plt.plot(x, norm.cdf(x, mu, sigma), 'black', linewidth=1.5)
plt.xlabel('$t$', fontsize=14)
plt.ylabel('$f(t)$', fontsize=14)
plt.grid(True)
plt.axis([-5, 5, -0.01, 1.01])
plt.show()
```

```python
# Plot discrete approximation of pdf
N = 1000
x = np.random.normal(0, 1, N)

# Plot histogram
plt.hist(x, bins=np.arange(-5, 5, 0.4), density=True, edgecolor='black', facecolor='none', linewidth=1)

plt.xlabel('$x$ [bar]', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.grid(False)
plt.axis([-5, 5, 0, 0.42])
plt.show()
```

```python
# Compute an plot empirical CDF
dx = 0.5
t = np.arange(-5,5+dx,dx)
N = len(x)

# Initialize the cumulative distribution function array
F = np.zeros(len(t))

# Compute the empirical CDF
for i in range(len(t)):
    for j in range(N):
        if x[j] <= t[i]:
            F[i] += 1
F = F/N # Normalize
```

```python
# Plot the empirical CDF
plt.step(t, F, color='black', where='mid')
plt.xlabel('$x$ [bar]')
plt.ylabel('$F(x)$')
plt.axis([-5, 5, -0.01, 1.01])
plt.grid(True)
plt.show()
```
