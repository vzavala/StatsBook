```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.interpolate import interp1d
```

```python
# PDF
x = np.array([-1, 0, 1, 2])
f = np.array([0, 0.4, 0.6, 0])
n = len(x)
```

```python
# Compute CDF using PDF
t = np.linspace(-1, 2, 1000)
F = np.zeros_like(t)

for i in range(len(t)):
    for j in range(len(x)):
        F[i] += (x[j] <= t[i]) * f[j]
```

```python
# Visualize PDF
plt.plot(0*np.ones(100), np.linspace(0, f[1], 100), 'k--')
plt.plot(1*np.ones(100), np.linspace(0, f[2], 100), 'k--')
plt.scatter(x, f, marker='o', s=20, color='white', edgecolor='black')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.axis([-1, 2, -0.02, 1.02])
plt.grid(True)
plt.show()
```

```python
# Visualize PDF
plt.step(t, F, 'k--', where='mid')

# Interpolate CDF
interp_cdf = interp1d(t, F, kind='nearest')

# Scatter plot
for j in range(n):
    plt.scatter(x[j], interp_cdf(x[j]), marker='o', facecolors='w', edgecolors='black')

plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.axis([-1, 2, -0.02, 1.02])
plt.grid(True)
plt.show()
```
