```python
# Victor Z and J E Uma√±a
# UW-Madison, 2024
# pdf and cdf of a Bernoulli RV

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# pdf
p = 0.3
x = np.array([-1, 0, 1, 2])
f = np.array([0, (1-p), p, 0])
n = len(x)

# compute cdf
t = np.linspace(-1, 2, 100)
F = np.zeros_like(t)
for i in range(len(t)):
    for j in range(len(x)):
        F[i] += (x[j] <= t[i]) * f[j]

# generate samples
S = 100
X = binom.rvs(1, p, size=S)

plt.figure(figsize=(10,8))

# plot samples
plt.subplot(2, 2, (1, 2))
plt.plot(X, 'ko', markerfacecolor='w', markersize=4)
plt.xlabel(r'Outcome $\omega$', fontsize=14)
plt.ylabel(r'$x_\omega$', fontsize=14)
plt.grid(color='gray',axis='both',alpha=1)
plt.yticks([-0.5, 0, 0.5, 1, 1.5])
plt.axis([0, S, -0.5, 1.5])

# visualize pdf
plt.subplot(2, 2, 3)
plt.plot(0 * np.ones(100), np.linspace(0, f[1], 100), 'k--', zorder=2)
plt.plot(1 * np.ones(100), np.linspace(0, f[2], 100), 'k--', zorder=2)
plt.scatter(x, f, marker='o', facecolors='w', edgecolors='k', zorder=3)
plt.grid(color='gray',axis='both',alpha=1, zorder=1)
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$f(x)$', fontsize=14)
plt.axis([-0.5, 1.5, -0.02, 1.02])

# visualize cdf
plt.subplot(2, 2, 4)
plt.grid(color='gray',axis='both',alpha=1, zorder=1)
plt.step(t, F, 'k-', zorder=2, where='post')
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$F(x)$', fontsize=14)
plt.axis([-0.5, 1.5, -0.02, 1.02])
for j in range(n):
    plt.scatter(x[j], np.interp(x[j], t, F), marker='o', facecolors='w', edgecolors='k', zorder=3)

plt.show()
```
