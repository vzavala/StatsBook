```python
# Victor Z, J E Umaa
# UW-Madison, 2024
# compare chi-squared

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, chi2, norm

# Clear previous plots
plt.close('all')

# Define parameters
beta = [2, 2, 2, 2]
alpha = [1/2, 2/2, 3/2, 4/2]
colors = ["blue", "red", "green", "black"]
x = np.linspace(0, 4, 1000)

# Plot gamma PDF and CDF
plt.figure(figsize=(10,8))

plt.subplot(2, 2, 1)
for k in range(len(beta)):
    plt.plot(x, gamma.pdf(x, alpha[k], scale=beta[k]), color=colors[k], linewidth=1.5)
plt.grid(True)
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$f(x)$', fontsize=12)
plt.title('')
plt.axis([0, 4, 0, 1])

plt.subplot(2, 2, 2)
for k in range(len(beta)):
    plt.plot(x, gamma.cdf(x, alpha[k], scale=beta[k]), color=colors[k], linewidth=1.5)
plt.grid(True)
lgd = plt.legend(['$r=1$', '$r=2$', '$r=3$', '$r=4$'], loc='lower right', fontsize=8, framealpha=1, edgecolor='k')
plt.xlabel('$x$', fontsize=12)
plt.ylabel('$F(x)$', fontsize=12)
plt.axis([0, 4, 0, 1])
plt.title('')
```

```python
# Plot quantile functions for chi2(1) and normal(0,1)
alpha = np.linspace(0.01, 1, 100)

plt.figure(figsize=(10,8))

plt.subplot(2, 2, 1)
plt.plot(alpha, chi2.ppf(1-alpha, df=1), 'b-', linewidth=1.5)
plt.grid(True)
lgd = plt.legend(['$\chi^2(1)$'], loc='upper right', fontsize=8, framealpha=1, edgecolor='k')
plt.ylabel('$Q(1-\\alpha)$', fontsize=12)
plt.xlabel('$\\alpha$', fontsize=12)
plt.axis([0, 1, 0, 8])
plt.title('')

plt.subplot(2, 2, 2)
plt.plot(alpha, norm.ppf(1-alpha/2)**2, 'b-', linewidth=1.5)
plt.grid(True)
lgd = plt.legend(['$N(0,1)$'], loc='upper right', fontsize=8, framealpha=1, edgecolor='k')
plt.ylabel('$Q(1-\\alpha/2)^2$', fontsize=12)
plt.xlabel('$\\alpha$', fontsize=12)
plt.axis([0, 1, 0, 8])
plt.title('')

plt.show()
```
