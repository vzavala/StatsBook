```python
# Victor Z, J E Umaña
# UW-Madison, 2024
# compare exponential RVs

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Clear previous plots
plt.close('all')

# Parameter values
beta = [1, 2, 4]
w = [0.5, 1, 1.5, 2]
x = np.linspace(0, 4, 1000)

# Generate realizations
S = 1000
np.random.seed(0)
X = np.random.exponential(scale=4, size=S)

# Plot
plt.figure(figsize=(10,8))

plt.subplot(2, 2, (1, 2))
plt.plot(X, 'ko', markerfacecolor='w', markersize=4)
plt.xlabel('Outcome $\omega$', fontsize=14)
plt.ylabel('$x_\omega$', fontsize=14)
plt.axis([0, 1000, 0, 36])
plt.grid(True)

plt.subplot(2, 2, 3)
for k in range(len(beta)):
    plt.plot(x, expon.pdf(x, scale=beta[k]), 'k', linewidth=w[k])
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.axis([0, 4, 0, 1])
plt.title('')

plt.subplot(2, 2, 4)
for k in range(len(beta)):
    plt.plot(x, expon.cdf(x, scale=beta[k]), 'k', linewidth=w[k])
plt.plot(x, np.ones_like(x) * 0.63, 'k--')
plt.grid(True)
lgd = plt.legend(['$ \\beta=1 $', '$  \\beta =2 $', '$ \\beta=4 $'], loc='lower right', fontsize=8, framealpha=1, edgecolor='k')
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$F(x)$', fontsize=14)
plt.axis([0, 4, 0, 1])
plt.title('')

plt.show()
```
