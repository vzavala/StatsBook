```python
# Victor Z, J E Umaña
# UW-Madison, 2024
# compare gamma RVs

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Clear previous plots
plt.close('all')

# Parameters
beta = [1, 1, 0.5, 0.5]
alpha = [1, 2, 1, 2]
w = [0.5, 1, 1.5, 2]
x = np.linspace(0, 4, 1000)

# Generate realizations
S = 1000
np.random.seed(0)

# Plot
plt.figure(figsize=(10,8))

plt.subplot(2, 2, 1)
X = np.random.gamma(1, scale=1, size=S)
plt.plot(X, 'ko', markerfacecolor='w', markersize=4)
plt.xlabel('Outcome $\omega$', fontsize=14)
plt.ylabel('$x_\omega$', fontsize=14)
plt.grid(True)
plt.axis([0, 1000, 0, 10])

plt.subplot(2, 2, 2)
X = np.random.gamma(2, scale=1, size=S)
plt.plot(X, 'ko', markerfacecolor='w', markersize=4)
plt.xlabel('Outcome $\omega$', fontsize=14)
plt.ylabel('$x_\omega$', fontsize=14)
plt.grid(True)
plt.axis([0, 1000, 0, 10])

plt.subplot(2, 2, 3)
for k in range(len(beta)):
    plt.plot(x, gamma.pdf(x, a=alpha[k], scale=beta[k]), 'k', linewidth=w[k],
             label=f'$\\beta={beta[k]},\\alpha={alpha[k]}$')
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
lgd = plt.legend(loc='upper right', fontsize=8, framealpha=1, edgecolor='k')
plt.axis([0, 4, 0, 2])
lgd.set_title('Gamma parameters', prop={'size': 10})

plt.subplot(2, 2, 4)
for k in range(len(beta)):
    plt.plot(x, gamma.cdf(x, a=alpha[k], scale=beta[k]), 'k', linewidth=w[k],
             label=f'$\\beta={beta[k]},\\alpha={alpha[k]}$')
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$F(x)$', fontsize=14)
plt.axis([0, 4, 0, 1])
plt.title('')

plt.show()
```
