```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parameters
mu = [0, 0, 1]
sigma = [np.sqrt(1/5), np.sqrt(1), np.sqrt(1/10)]
w = [0.5, 1.5, 2.5]
x = np.linspace(0.001, 5, 1000)

# Generate realizations
S = 1000
np.random.seed(0)
X = np.random.lognormal(mean=0, sigma=2, size=S)
```

```python
# Plotting
plt.figure(figsize=(10, 8))

# Subplot 1: Realizations (linear scale)
plt.subplot(2, 2, 1)
plt.plot(X, 'ko', markerfacecolor='w', markersize=4)
plt.xlabel('Outcome $\\omega$', fontsize=14)
plt.ylabel('$x_\\omega$', fontsize=14)
plt.grid(True)

# Subplot 2: Realizations (log scale)
plt.subplot(2, 2, 2)
plt.semilogy(X, 'ko', markerfacecolor='w', markersize=4)
plt.xlabel('Outcome $\\omega$', fontsize=14)
plt.ylabel('$x_\\omega$', fontsize=14)
plt.grid(True)

# Subplot 3: PDF
plt.subplot(2, 2, 3)
for k in range(len(mu)):
    s = sigma[k]
    scale = np.exp(mu[k])
    plt.plot(x, lognorm.pdf(x, s, scale=scale), 'k-', linewidth=w[k])
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)

# Subplot 4: CDF
plt.subplot(2, 2, 4)
for k in range(len(mu)):
    s = sigma[k]
    scale = np.exp(mu[k])
    plt.plot(x, lognorm.cdf(x, s, scale=scale), 'k-', linewidth=w[k])
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$F(x)$', fontsize=14)
plt.legend([
    '$\\mu=0,\\sigma^2=0.2$',
    '$\\mu=0,\\sigma^2=1$',
    '$\\mu=1,\\sigma^2=0.1$'
], loc='lower right', fontsize=10)

plt.tight_layout()
plt.show()
```
