```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters
a, b = -1, 1
x = np.linspace(-2, 2)

# Generate realizations
S = 1000
np.random.seed(0)
X = np.random.uniform(a, b, S)
```

```python
# Plotting
plt.figure(figsize=(10, 8))

# Subplot 1: Realizations
plt.subplot(2, 2, (1, 2))
plt.plot(X, 'ko', markerfacecolor='w', markersize=4)
plt.xlabel('Outcome $\\omega$', fontsize=14)
plt.ylabel('$x_\\omega$', fontsize=14)
plt.grid(True)
plt.yticks([-2, -1, 0, 1, 2])
plt.axis([0, S, -2, 2])

# Subplot 2: PDF
plt.subplot(2, 2, 3)
plt.step(x, uniform.pdf(x, loc = a, scale = 2), where='mid', color='black', linewidth=1.5)#plt.stairs(x, uniform.pdf(x, a, b), color='black', linewidth=1.5)
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)

# Subplot 3: CDF
plt.subplot(2, 2, 4)
plt.plot(x, uniform.cdf(x, loc = a, scale = 2), color='black', linewidth=1.5)
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$F(x)$', fontsize=14)
plt.title('')

plt.tight_layout()
plt.show()
```
