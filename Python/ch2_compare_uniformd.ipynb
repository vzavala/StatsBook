```python
import numpy as np
import matplotlib.pyplot as plt

# PDF
a, b = -1, 1
x = np.arange(a, b + 1)
n = len(x)
f = np.ones(n) * (1 / n)

# Generate samples
S = 100
np.random.seed(0)
X = np.random.randint(1, 4, S) - 2
```

```python
# Plotting
plt.figure(figsize=(10, 8))

# Subplot 1: Realizations
plt.subplot(2, 2, (1, 2))
plt.plot(X, 'ko', markerfacecolor='w', markersize=4)
plt.xlabel('Outcome $\\omega$', fontsize=14)
plt.ylabel('$x_\\omega$', fontsize=14)
plt.grid(True)
plt.yticks([-1, 0, 1])
plt.axis([0, S, -1.5, 1.5])

# Compute CDF using PDF
t = np.linspace(-2, 2, 1000)
F = np.zeros_like(t)
for i, val in enumerate(t):
    F[i] = np.sum(f[x <= val])

# Subplot 2: PDF
plt.subplot(2, 2, 3)
for j in range(n):
    plt.plot(x[j] * np.ones(100), np.linspace(0, f[j], 100), 'k--')
plt.scatter(x, f, marker='o', facecolors='w', edgecolors='black')
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.axis([-1.5, 1.5, -0.02, 0.52])

# Subplot 3: CDF
plt.subplot(2, 2, 4)
for j in range(n):
    Fn = (np.floor(x[j]) - a + 1) / n
    plt.step([x[j], x[j] + 1], [Fn, Fn], where='post', color='black')
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$F(x)$', fontsize=14)
plt.axis([-1.5, 1.5, -0.02, 1.02])

plt.tight_layout()
plt.show()
```
