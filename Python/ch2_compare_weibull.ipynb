```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# define parameters
beta = [1, 1, 2, 2]
xi = [1, 2, 1, 2]
w = [0.5, 1, 1.5, 2]
x = np.linspace(0, 4, 1000)
```

```python
plt.figure(figsize=(10, 8))

# Subplot 1: PDF
plt.subplot(2, 2, 1)
for k in range(len(beta)):
    plt.plot(x, weibull_min.pdf(x, beta[k], scale=xi[k]), color='black', linewidth=w[k])
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.title('')
plt.legend(['$\\beta=1, \\xi=1$', '$\\beta=1, \\xi=2$', '$\\beta=2, \\xi=1$', '$\\beta=2, \\xi=2$'], fontsize=12)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)

# Subplot 2: CDF
plt.subplot(2, 2, 2)
for k in range(len(beta)):
    plt.plot(x, weibull_min.cdf(x, beta[k], scale=xi[k]), color='black', linewidth=w[k])
plt.plot(x, np.ones_like(x) * 0.63, 'k--')
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$F(x)$', fontsize=14)
plt.title('')
plt.legend(['$\\beta=1, \\xi=1$', '$\\beta=1, \\xi=2$', '$\\beta=2, \\xi=1$', '$\\beta=2, \\xi=2$'], fontsize=12, loc='lower right')
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$F(x)$', fontsize=14)

plt.tight_layout()
plt.show()
```
