```python
# Victor Z, J E Uma√±a
# UW-Madison, 2024
# example binomial

import numpy as np
from scipy.stats import binom

# Set seed for reproducibility
np.random.seed(0)

# Probability of outcome
p = 0.5

# Sequence length
n = 10

# Find 2 failures
x = 2
P = binom.pmf(x, n, p)
print(f"Probability of {x} failures: {np.round(P,4)}")

# Find 10 failures
x = 10
P = binom.pmf(x, n, p)
print(f"Probability of {x} failures: {np.round(P,8)}")

# Find average
x = n * p
P = binom.pmf(x, n, p)
print(f"Average number of failures : {int(x)}")
print(f"Probability of average number of failures ({int(x)}): {np.round(P,4)}")

# Find 0 failures
x = 0
P = binom.pmf(x, n, p)
print(f"Probability of {x} failures: {np.round(P,8)}")
```
