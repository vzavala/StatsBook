```python
# Victor Z, J E Umaa
# UW-Madison, 2024
# exponential failure adhesive material

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Clear previous plots
plt.close('all')

# Average failure time
eta = 3 * 12  # months

# Generate outcomes
np.random.seed(0)
S = 1000
x = np.random.exponential(scale=eta, size=S)

# Plot outcomes
plt.figure(figsize=(10, 12))

plt.subplot(3, 1, 1)
plt.plot(x, 'ko', markerfacecolor='w', markersize=4)
plt.xlabel('Outcome $\omega$', fontsize=14)
plt.ylabel('$x_\omega$ [months]', fontsize=14)
plt.grid(True)
plt.axis([0, 1000, 0, 300])

# Empirical & theoretical pdf
plt.subplot(3, 1, 2)
xx = np.linspace(0, 160)
plt.plot(xx, expon.pdf(xx, scale=eta), 'k', linewidth=1.5, zorder=2)
plt.hist(x, bins=np.arange(0, 161, 10), density=True, edgecolor='black', facecolor='none', linewidth=1, zorder=2)
plt.xlabel('$x$ [months]', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.grid(True, zorder=1)
plt.axis([0, 160, 0, 0.03])

# Theoretical cdf
plt.subplot(3, 1, 3)
plt.plot(xx, expon.cdf(xx, scale=eta), 'k', linewidth=1.5)
plt.xlabel('$x$ [months]', fontsize=14)
plt.ylabel('$F(x)$', fontsize=14)
plt.grid(True)
plt.axis([0, 160, 0, 1])
```

```python
# Probability of surviving after 12 months
probability_surviving = 1 - expon.cdf(12, scale=eta)
print(f"Probability of surviving after 12 months: {np.round(probability_surviving,4)}")
```
