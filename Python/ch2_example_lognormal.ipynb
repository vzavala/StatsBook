```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm
from scipy.special import erfinv

# generate data
mu = np.log(200)
logq = np.log(400)
alpha = 0.9
sigma = (logq - mu) / (np.sqrt(2) * erfinv(2 * alpha - 1))
x = np.linspace(10, 1000, 1000)
expF = [0.1, 0.5, 0.9]
expx = [95, 200, 411]
```

```python
plt.figure(1)

# visualize cdf
plt.subplot(2, 1, 1)
plt.semilogx(x, lognorm.cdf(x, s=sigma, scale=np.exp(mu)), color = 'black', linewidth=1.5)
plt.plot(expx, expF, 'o', markerfacecolor='white', markeredgecolor='black', markersize=10, linewidth=1.2)
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$F(x)$', fontsize=14)
plt.legend(['Theoretical', 'Experimental Data'], loc='upper left', fontsize=10)

# visualize pdf
plt.subplot(2, 1, 2)
plt.semilogx(x, lognorm.pdf(x, s=sigma, scale=np.exp(mu)), color = 'black', linewidth=1.5)
plt.grid(True)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# compute probability that 250 < X < 400
P = lognorm.cdf(400, s=sigma, scale=np.exp(mu)) - lognorm.cdf(250, s=sigma, scale=np.exp(mu))
print(f"Probability that 250 < X < 400: {P}")
```
