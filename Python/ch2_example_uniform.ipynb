```python
import numpy as np
import matplotlib.pyplot as plt

# for reproducibility
np.random.seed(0)

z = 52  # weeks
t = np.arange(0, z+1)  # time vector
n = len(t)  # number of times
N = 3  # number of samples
```

```python
# generate random failure sequences
ny = 0
plt.figure()

for k in range(1, N+1):

    y = np.random.randint(0, 2, n)  # generate random sequence of 0s and 1s
    ny += np.sum(y)

    plt.subplot(2, 1, 1)
    for j in range(len(y)):
        if y[j] == 1:  # failure
            plt.scatter(t[j], k, edgecolor='black', facecolor='black')
        else:  # normal
            plt.scatter(t[j], k, edgecolor='black', facecolor='none')

plt.grid(True)
plt.axis([min(t), max(t), 0, 4])
plt.gca().set_yticklabels(['', 'Year 1', 'Year 2', 'Year 3', ''])
plt.xlabel('Time [weeks]', fontsize=14)
plt.box(True)

plt.tight_layout()
plt.show()
```

```python
failure_rate = ny / (N * z)
print(f"Failure rate: {failure_rate}")
```
