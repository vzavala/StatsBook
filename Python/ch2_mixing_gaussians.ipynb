```python
# Victor Z, J E Uma√±a
# UW-Madison, 2024
# compare cdfs for standard normal and normal for mixing problem

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Clear previous plots
plt.close('all')

# Define x range
x = np.linspace(-4, 4)
n = len(x)

# Parameters for normal distribution
mu = 30
sig = 1

# Probability level
alpha = 0.977

# Quantile for N(0,1)
q1 = norm.ppf(alpha, 0, 1)

# Quantile for N(mu, sigma)
q1p = mu + q1 * sig
print(q1p)

# Confirming the quantile for N(mu, sigma)
q1p = norm.ppf(alpha, mu, sig)
print(q1p)
```

```python
# Plot quantile function for N(0,1)
plt.figure(figsize=(10,8))
plt.subplot(2, 2, 1)
plt.plot(x, norm.cdf(x, 0, 1), 'black', linewidth=1.5)
plt.grid(True)
plt.plot(x, alpha * np.ones_like(x), 'k--', lw=1)
plt.plot(q1 * np.ones(n), np.linspace(0, 1), 'k--', lw=1)
plt.text(0.98, 0.05, '$z_\\alpha\\rightarrow$', fontsize=14)
plt.xlabel('$Q(\\alpha)$', fontsize=14)
plt.ylabel('$\\alpha$', fontsize=14)
plt.axis([-4, 4, 0, 1.02])

# Plot quantile function for N(mu, sigma)
x = np.linspace(mu - 4 * sig, mu + 4 * sig)
plt.subplot(2, 2, 2)
plt.plot(x, norm.cdf(x, mu, sig), 'black', linewidth=1.5)
plt.grid(True)
plt.plot(x, alpha * np.ones_like(x), 'k--', lw=1)
plt.plot(q1p * np.ones(n), np.linspace(0, 1), 'k--', lw=1)
plt.text(29.6, 0.05, '$\\mu+z_\\alpha\\cdot \\sigma\\rightarrow$', fontsize=14)
plt.xlabel('$Q(\\alpha)$', fontsize=14)
plt.ylabel('$\\alpha$', fontsize=14)
plt.axis([26, 34, 0, 1.02])
```

```python

# Compute 90% confidence interval
plt.figure(figsize=(10,8))
x = np.linspace(mu - 4 * sig, mu + 4 * sig)
n = len(x)

# Probability level
alpha = 0.10

# Quantile for N(0,1)
q = norm.ppf(1 - alpha / 2, 0, 1)

# Intervals
qup = mu + q * sig
qlow = mu - q * sig

plt.plot(x, norm.pdf(x, mu, sig), 'black', linewidth=1.5)
plt.grid(True)
plt.plot(qup * np.ones(n), np.linspace(0, 0.4), 'k--', linewidth=1.5)
plt.plot(qlow * np.ones(n), np.linspace(0, 0.4), 'k--', linewidth=1.5)
plt.text(29.825, 0.01, '$\\Longleftrightarrow$', fontsize=14)
plt.text(qup - 0.4, 0.405, '$\\mu+ z_{\\alpha/2}\\sigma$', fontsize=14)
plt.text(qlow - 0.4, 0.405, '$\\mu- z_{\\alpha/2}\\sigma$', fontsize=14)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.axis([26, 34, 0, 0.4])
plt.show()
```
