```python
import numpy as np
import matplotlib.pyplot as plt

# Data set 1
mu = [0, 0]
Sigma = [[1, -0.9], [-0.9, 1]]  # anticorrelated

np.random.seed(0)
nsample = 1000
sample = np.random.multivariate_normal(mu, Sigma, nsample)

plt.figure(1)
plt.subplot(1, 1, 1)
plt.plot(sample[:, 0], sample[:, 1], 'ko', markerfacecolor='w')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.grid(True)
plt.xticks([])
plt.yticks([])
```

```python
# Data set 2
mu = [0, 0]
Sigma = [[1, 0], [0, 1]]  # uncorrelated

np.random.seed(0)
sample = np.random.multivariate_normal(mu, Sigma, nsample)

plt.subplot(1, 1, 1)
plt.plot(sample[:, 0], sample[:, 1], 'ko', markerfacecolor='w')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.grid(True)
plt.xticks([])
plt.yticks([])
```

```python
# Data set 3
nsample = 200
x1 = np.linspace(0, 5, nsample)
x2 = x1 ** 2 + np.random.normal(0, 1, nsample)

plt.subplot(1, 1, 1)
plt.plot(x2, x1, 'ko', markerfacecolor='w')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.grid(True)
plt.xticks([])
plt.yticks([])
```

```python
# Data set 4
np.random.seed(0)
x1, x2 = np.meshgrid(np.arange(0, 11, 0.5), np.arange(0, 11, 0.5))
x3 = 2 * x1**2 + 2 * x2**2 + 20 * np.random.normal(0, 1, x1.shape)

fig = plt.subplot(1, 1, 1, projection='3d')
fig.scatter(x1, x2, x3, c='k', marker='o', facecolors='w')
fig.set_xlabel('$x_1$', fontsize=14)
fig.set_ylabel('$x_2$', fontsize=14)
fig.set_zlabel('$x_3$', fontsize=14)
fig.grid(True)
fig.set_xticks([])
fig.set_yticks([])
fig.set_zticks([])

plt.tight_layout()
plt.show()
```
