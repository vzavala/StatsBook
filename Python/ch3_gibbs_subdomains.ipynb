```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
path = '/content/drive/MyDrive/StatsBookCode/Python/'
filename = path+'Data/gibbs_hightemp_class.dat'
datahigh = np.loadtxt(filename)

data = np.array(datahigh)
datan = data.copy()
n = len(datan)

# Corrupt data with noise to hide pattern
np.random.seed(1)
P = datan[:, 1] + np.random.normal(0, 1, n)
C = datan[:, 2] + np.random.normal(0, 0.05, n)
```

```python
# Plot subdomains
plt.figure(figsize=(10, 8))

# Draw domain for P in [100 200]
plt.subplot(2, 2, 1)
plt.fill_between([100, 200], 0, 1, color='lightgrey')
plt.scatter(P, C, color='black', marker='o', facecolors='white')
plt.grid(True)
plt.xlabel('Pressure [bar]', fontsize=14)
plt.ylabel('Conversion [-]', fontsize=14)
plt.axis([50, 250, 0, 1])

# Draw domain for P in [100,200] and C in <= 0.5
plt.subplot(2, 2, 2)
plt.fill_between([100, 200], 0, 0.5, color='lightgrey')
plt.scatter(P, C, color='black', marker='o', facecolors='white')
plt.grid(True)
plt.xlabel('Pressure [bar]', fontsize=14)
plt.ylabel('Conversion [-]', fontsize=14)
plt.axis([50, 250, 0, 1])

# Draw domain for P <= 100 or P>=200 and C in >= 0.5
plt.subplot(2, 2, 3)
plt.fill_between([0, 100], 0.5, 1, color='lightgrey')
plt.fill_between([200, 250], 0.5, 1, color='lightgrey')
plt.scatter(P, C, color='black', marker='o', facecolors='white')
plt.grid(True)
plt.xlabel('Pressure [bar]', fontsize=14)
plt.ylabel('Conversion [-]', fontsize=14)
plt.axis([50, 250, 0, 1])

# Draw domain for P <= 100 and P>=200 and C in >= 0.5 (this is impossible)
plt.subplot(2, 2, 4)
plt.scatter(P, C, color='black', marker='o', facecolors='white')
plt.grid(True)
plt.xlabel('Pressure [bar]', fontsize=14)
plt.ylabel('Conversion [-]', fontsize=14)
plt.axis([50, 250, 0, 1])

plt.tight_layout()
plt.show()
```
