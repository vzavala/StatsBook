```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Generate observations for exponential distribution
# note there are discrepancies with Matlab due to use of different samples
np.random.seed(0)
S = 1000
x = np.random.exponential(scale=2, size=S)

# Compute log likelihood
beta = np.linspace(0.1, 10, 100)  # Avoid zero to prevent division by zero
ss = np.sum(x)
logL = -S * np.log(beta) - (1 / beta) * ss

# Estimate parameter
betahat = np.mean(x)
print(f'betahat: {betahat}')
```

```python
# Plotting
plt.figure(figsize=(10, 8))

# Scatter plot of observations
plt.subplot(2, 2, (1, 2))
plt.plot(x, 'o', markerfacecolor='w', markeredgecolor='black', markersize=4)
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$x_\omega$', fontsize=14)
plt.grid(True)

# Log likelihood plot
plt.subplot(2, 2, 3)
plt.plot(beta, logL, 'k-', linewidth=1.5)
plt.xlabel(r'$\beta$', fontsize=14)
plt.ylabel(r'$\log L(\beta)$', fontsize=14)
plt.grid(True)
plt.axis([0, 10, np.min(logL), np.max(logL)])

# Histogram and PDF plot
plt.subplot(2, 2, 4)
xgrid = np.linspace(0, 10, 100)
plt.hist(x, bins=np.arange(0, 10 + 1, 1), density=True, edgecolor='black', facecolor='none', linewidth=1)
plt.plot(xgrid, expon.pdf(xgrid, scale=betahat), 'k-', linewidth=1.5)
plt.xlabel(r'$x$', fontsize=14)
plt.ylabel(r'$f(x)$', fontsize=14)
plt.grid(True)

plt.tight_layout()
plt.show()
```
