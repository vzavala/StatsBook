```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to generate normal distribution data and compute parameters
# note there are discrepancies with Matlab due to use of different samples
def generate_and_plot(S, mu, sig, subplot_idx):
    np.random.seed(1)
    X = np.random.normal(mu, sig, S)

    # Moment matching estimates
    muhat = np.mean(X)
    sighat = np.std(X)

    print(f'For S={S}: muhat={muhat}, sighat={sighat}')

    # Plot histogram and PDF
    plt.subplot(2, 2, subplot_idx)
    plt.hist(X, bins=np.arange(min(X), max(X) + 0.5, 0.5), density=True, edgecolor='black', facecolor='none', linewidth=1)
    x = np.linspace(-10, 10, 100)
    plt.plot(x, norm.pdf(x, mu, sig), color='black', linewidth=1.5)
    plt.xlabel('$x$', fontsize=14)
    plt.ylabel('$f(x)$', fontsize=14)
    plt.grid(True)

    # Compute and plot empirical CDF
    plt.subplot(2, 2, subplot_idx + 1)
    sorted_X = np.sort(X)
    F = np.arange(1, S + 1) / S
    plt.step(sorted_X, F, where='post', color='black', linewidth=1)
    plt.plot(x, norm.cdf(x, mu, sig), color='black', linewidth=1.5)
    plt.xlabel('$x$', fontsize=14)
    plt.ylabel('$F(x)$', fontsize=14)
    plt.grid(True)
```

```python
# Try few data points
plt.figure(figsize=(10, 8))

generate_and_plot(10, 1, 2, 1)

# Try more data points
generate_and_plot(100, 1, 2, 3)

plt.tight_layout()
plt.show()
```
