```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random observations
np.random.seed(0)
S = 100
x = np.random.normal(0, 1, S)
xx = np.arange(1, S+1)

# Generate non-random observations
# Use mechanism to discard sample if above threshold
S2 = 200
xt = np.random.normal(0, 1, S2)
x2 = xt[xt <= 1][:S]

# Compute means for both
mx = np.full(S, np.mean(x))
mx2 = np.full(S, np.mean(x2))

# Compute bias
bias1 = np.mean(x - mx)
bias2 = np.mean(x2 - mx)

print(f'bias1: {bias1}')
print(f'bias2: {bias2}')
```

```python
# Display sequences
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(xx, mx, 'black', linewidth=1.5)
plt.plot(xx, x, 'o', markerfacecolor='w', markeredgecolor='black')
plt.grid(True)
plt.axis([0, S, -4, 4])
plt.xlabel('$\\omega$', fontsize=14)
plt.ylabel('$x_\\omega$', fontsize=14)

plt.subplot(2, 1, 2)
plt.plot(xx, mx, 'black', linewidth=1.5)
plt.plot(xx, x2, 'o', markerfacecolor='w', markeredgecolor='black')
plt.grid(True)
plt.axis([0, S, -4, 4])
plt.xlabel('$\\omega$', fontsize=14)
plt.ylabel('$x_\\omega$', fontsize=14)

plt.tight_layout()
plt.show()
```
