```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# span theta values
th = np.linspace(-3, 3, 100)

# construct prior function p(th)
sig0 = 1
th0 = 0
logpth = -np.log(np.sqrt(2 * np.pi * sig0**2)) - ((th - th0)**2) / (2 * sig0)

# construct p(y|th)
sig = 1
y = 2
x = 1

logpyth = -np.log(np.sqrt(2 * np.pi * sig**2)) - ((y - th * x)**2) / (2 * sig)

# construct p(th|y)
logpthy = logpyth + logpth
```

```python
# visualize prior and conditional in log space
plt.figure(1, figsize=(10, 10))
plt.subplot(3, 3, 1)
plt.plot(th, logpth, 'k', linewidth=1.5)
plt.xlabel(r'$\theta$', fontsize=16)
plt.ylabel(r'$\log f(\theta)$', fontsize=16)
plt.grid(True)

plt.subplot(3, 3, 2)
plt.plot(th, logpyth, 'k', linewidth=1.5)
plt.xlabel(r'$\theta$', fontsize=16)
plt.ylabel(r'$\log f(y|\theta)$', fontsize=16)
plt.grid(True)

plt.subplot(3, 3, 3)
plt.plot(th, logpthy, 'k', linewidth=1.5)
plt.xlabel(r'$\theta$', fontsize=16)
plt.ylabel(r'$\log f(\theta|y)$', fontsize=16)
plt.grid(True)

# visualize in original space
plt.subplot(3, 3, 4)
plt.plot(th, np.exp(logpth), 'k', linewidth=1.5)
plt.xlabel(r'$\theta$', fontsize=16)
plt.ylabel(r'$f(\theta)$', fontsize=16)
plt.grid(True)

plt.subplot(3, 3, 5)
plt.plot(th, np.exp(logpyth), 'k', linewidth=1.5)
plt.xlabel(r'$\theta$', fontsize=16)
plt.ylabel(r'$f(y|\theta)$', fontsize=16)
plt.grid(True)

plt.subplot(3, 3, 6)
plt.plot(th, np.exp(logpthy), 'k', linewidth=1.5)
plt.xlabel(r'$\theta$', fontsize=16)
plt.ylabel(r'$f(\theta|y)$', fontsize=16)
plt.grid(True)

plt.tight_layout()
plt.show()
```
