```python
# Victor Z and Leo Gonzalez
# UW-Madison, 2024
# examples of quadratic surfaces

import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# generate a surface with sharp minimum
H = np.array([[2, 0],
              [0, 2]])
print(np.linalg.eigvals(H))

X1 = np.arange(-1, 1+0.005, 0.005)
X1, X2 = np.meshgrid(X1, X1)
SSE = H[0, 0]*X1**2+2*H[0, 1]*X1*X2+H[1, 1]*X2**2

fig = plt.figure(figsize = [12, 12])
ax1 = fig.add_subplot(2, 2, 1, projection = '3d')
ax1.grid(color = 'gray', axis = 'both', alpha = 0.25)
ax1.set_axisbelow(True)
ax1.set_xlabel(r'$\theta_1$', fontsize = 14)
ax1.set_ylabel(r'$\theta_2$', fontsize = 14)
ax1.set_zlabel(r'$SSE(\theta)$', fontsize = 14)
ax1.xaxis.set_ticklabels([])
ax1.yaxis.set_ticklabels([])
ax1.zaxis.set_ticklabels([])
surface = ax1.plot_surface(X1, X2, SSE, facecolors = plt.cm.Greys(SSE/np.max(SSE)))
contour = ax1.contour(X1, X2, SSE, zdir = 'z', offset = np.min(SSE), cmap = 'Greys')
```
