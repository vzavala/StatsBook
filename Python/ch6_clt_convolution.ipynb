```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform
```

```python
# Generate time values from 0 to 10 with 100 points
t = np.linspace(0, 10, 100)

# Generate the probability density function (PDF) using uniform distribution
# Parameters: t (values), loc (mean), scale (standard deviation)
f = uniform.pdf(t, loc=4, scale=6-4)
```

```python
# Show that repetitive convolutions yield Gaussian
phi = np.copy(f)

plt.step(t, phi/max(phi), color='black', where='mid')
plt.ylabel('$\phi(x)$')
plt.xlabel('$x$')
plt.grid(True)
plt.show()
```

```python
# Perform repetitive convolutions and plot
for k in range(3):
    phi = np.convolve(phi, f, mode='same')
    plt.subplot(2, 2, k + 1)
    plt.plot(t, phi / np.max(phi), color='black', linewidth=1.5)
    plt.grid(True)
    plt.ylabel('$\phi(x)$', fontsize=14)
    plt.xlabel('$x$', fontsize=14)

plt.show()
```
