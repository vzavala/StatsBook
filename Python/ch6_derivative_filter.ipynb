```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Generate derivative filter coefficients
w = np.array([1, 0, -1])

# Generate signal
N = 100
t = np.linspace(-10, 10, N)
f = norm.cdf(t)

plt.figure()

# Plot original signal
plt.subplot(2, 2, 1)
plt.plot(t, f, 'k', linewidth=1.5)
plt.grid(True)
plt.axis([-10, 10, 0, 1.05])
plt.ylabel('$f(t)$', fontsize=14)
plt.xlabel('$t$', fontsize=14)

# Perform convolution
phi = np.convolve(f, w, mode='same')

# Plot filtered signal
plt.subplot(2, 2, 2)
plt.plot(t, phi, 'k', linewidth=1.5)
plt.axis([-10, 10, 0, 0.16])
plt.grid(True)
plt.ylabel('$f*g(t)$', fontsize=14)
plt.xlabel('$t$', fontsize=14)

plt.tight_layout()
plt.show()
```
