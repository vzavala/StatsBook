```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create 2D periodic signals
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)

# Color space
c = np.linspace(0.99, 0.0, 100)
C = np.column_stack((c, c, c))
```

```python
# Create low freq signal
om1 = 10
f1 = np.sin(2 * np.pi * om1 * (X + Y))

fig = plt.figure(figsize=(12, 12))

# Plot low freq signal
ax1 = fig.add_subplot(3, 2, 3, projection='3d')
s1 = ax1.plot_surface(X, Y, f1, cmap='gray')
ax1.set_yticklabels([])
ax1.set_xticklabels([])
ax1.set_zticklabels([])
ax1.view_init(elev=30., azim=10)
ax1.set_ylabel(r'$x_2$', fontsize=14)
ax1.set_xlabel(r'$x_1$', fontsize=14)
ax1.set_zlabel(r'$f_1(x_1,x_2)$', fontsize=14)

ax2 = fig.add_subplot(3, 2, 4)
im1 = ax2.imshow(f1, cmap='gray', aspect='auto')
ax2.set_yticklabels([])
ax2.set_xticklabels([])
ax2.set_ylabel(r'$x_2$', fontsize=14)
ax2.set_xlabel(r'$x_1$', fontsize=14)

plt.show()
```

```python
# Create high freq signal
om2 = 60
f2 = np.sin(2 * np.pi * om2 * (X + Y))

fig = plt.figure(figsize=(12, 12))

# Plot high freq signal
ax3 = fig.add_subplot(3, 2, 5, projection='3d')
s2 = ax3.plot_surface(X, Y, f2, cmap='gray')
ax3.set_yticklabels([])
ax3.set_xticklabels([])
ax3.set_zticklabels([])
ax3.view_init(elev=30., azim=10)
ax3.set_ylabel(r'$x_2$', fontsize=14)
ax3.set_xlabel(r'$x_1$', fontsize=14)
ax3.set_zlabel(r'$f_2(x_1,x_2)$', fontsize=14)

ax4 = fig.add_subplot(3, 2, 6)
im2 = ax4.imshow(f2, cmap='gray', aspect='auto')
ax4.set_yticklabels([])
ax4.set_xticklabels([])
ax4.set_ylabel(r'$x_2$', fontsize=14)
ax4.set_xlabel(r'$x_1$', fontsize=14)

plt.show()
```

```python
# Total signal
f = f1 + f2

fig = plt.figure(figsize=(12, 12))

# Plot total signal
ax5 = fig.add_subplot(3, 2, 1, projection='3d')
s3 = ax5.plot_surface(X, Y, f, cmap='gray')
ax5.set_yticklabels([])
ax5.set_xticklabels([])
ax5.set_zticklabels([])
ax5.view_init(elev=30., azim=10)
ax5.set_ylabel(r'$x_2$', fontsize=14)
ax5.set_xlabel(r'$x_1$', fontsize=14)
ax5.set_zlabel(r'$f(x_1,x_2)$', fontsize=14)

ax6 = fig.add_subplot(3, 2, 2)
im3 = ax6.imshow(f, cmap='gray', aspect='auto')
ax6.set_yticklabels([])
ax6.set_xticklabels([])
ax6.set_ylabel(r'$x_2$', fontsize=14)
ax6.set_xlabel(r'$x_1$', fontsize=14)

plt.show()
```
