```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm
```

```python
# Parameters
L = 1000  # length of signal
Fs = 100  # Hz (sampling frequency)
T = 1 / Fs  # sec (sampling period)
t = np.arange(L) * T  # time vector
om = Fs * np.arange(L // 2 + 1) / L  # frequency vector
```

```python
# Check moving average
f = uniform.pdf(t, 4, 2)

fh = np.fft.fft(f)  # compute DFT
m = np.abs(fh)  # magnitude
mf = m[:L // 2 + 1]  # eliminate symmetry
mf[1:-1] = 2 * mf[1:-1]
```

```python
# Visualize signal in time and frequency domain
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(t, f, 'k', linewidth=1.5)
plt.grid(True)
plt.ylabel('$g(t)$', fontsize=14)
plt.xlabel('$t$', fontsize=14)

plt.subplot(2, 2, 2)
plt.plot(om, mf, 'k', linewidth=1.5)
plt.grid(True)
plt.ylabel('$|\\hat{g}(\\omega)|$', fontsize=14)
plt.xlabel('$\\omega$', fontsize=14)
plt.axis([0, 5, 0, 170])

# Get Gaussian signal
f = norm.pdf(t, 5, 0.4)

fh = np.fft.fft(f)  # compute DFT
m = np.abs(fh)  # magnitude
mf = m[:L // 2 + 1]  # eliminate symmetry
mf[1:-1] = 2 * mf[1:-1]

# Visualize signal in time and frequency domain
plt.subplot(2, 2, 3)
plt.plot(t, f, 'k', linewidth=1.5)
plt.grid(True)
plt.ylabel('$g(t)$', fontsize=14)
plt.xlabel('$t$', fontsize=14)

plt.subplot(2, 2, 4)
plt.plot(om, mf, 'k', linewidth=1.5)
plt.grid(True)
plt.ylabel('$|\\hat{g}(\\omega)|$', fontsize=14)
plt.xlabel('$\\omega$', fontsize=14)
plt.axis([0, 3, 0, max(mf)])

# Get another Gaussian signal
f = norm.pdf(t, 5, 1)

fh = np.fft.fft(f)  # compute DFT
m = np.abs(fh)  # magnitude
mf = m[:L // 2 + 1]  # eliminate symmetry
mf[1:-1] = 2 * mf[1:-1]

# Visualize signal in time and frequency domain
plt.subplot(2, 2, 3)
plt.plot(t, f, 'k--', linewidth=1.5)
plt.grid(True)
plt.ylabel('$g(t)$', fontsize=14)
plt.xlabel('$t$', fontsize=14)

plt.subplot(2, 2, 4)
plt.plot(om, mf, 'k--', linewidth=1.5)
plt.grid(True)
plt.ylabel('$|\\hat{g}(\\omega)|$', fontsize=14)
plt.xlabel('$\\omega$', fontsize=14)
plt.axis([0, 3, 0, 170])

plt.tight_layout()
plt.show()
```
