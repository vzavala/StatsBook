```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from matplotlib.colors import LinearSegmentedColormap
```

```python
# Specify mean and covariance
mu = [0, 0]
Sigma = [[2, 0], [0, 2]]  # No correlation
```

```python
# Evaluate probability density function in domain (-3,3) and (-3,3)
nmesh = 10
x1 = np.linspace(-3, 3, nmesh)
x2 = np.linspace(-3, 3, nmesh)
X1, X2 = np.meshgrid(x1, x2)  # Create mesh
```

```python
# Flatten the mesh and stack them for evaluation
pos = np.dstack((X1, X2))
rv = multivariate_normal(mu, Sigma)
F = rv.pdf(pos)

# Scale the PDF for visualization
Fc = np.round(100 * F)
```

```python
# Create a custom colormap from black to white
cmap = LinearSegmentedColormap.from_list('black_white', [(1, 1, 1), (0.5, 0.5, 0.5), (0, 0, 0)], N=64)

# Plotting
plt.figure(figsize=(6, 6))
plt.imshow(100 * F, cmap=cmap, interpolation='nearest', origin='lower')

plt.xticks([])
plt.yticks([])
plt.gca().tick_params(labelsize=14)
plt.show()
```
